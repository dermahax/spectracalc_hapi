## Evaluation methane lines.
- o) Overview
- a) 3260 nm (bisher)
- b) 3270 nm (IPM)
- c) 3245 nm (Total)

**Gas-cells:**
- Standard pressure and Temperature (296K), 
- 10 ppmm CH4
- 1 %m (i.e. 10 000 ppmm) H20

```python
%%capture
%load_ext autoreload
%autoreload 2
from classes import Spectra, Observer, HiddenPrints
```

```python
methane_cell = {'gas_name'  : "CH4",
                'VMR'       : 100E-6} # 100 ppm * 0.1 m
water_cell = {'gas_name'  : "H2O",   
              'VMR'       : 1E-1}  # 10% * 0.1 m
```

### o) Overview

```python
%%capture
#%% 1. create a spectrum
my_spectrum_overview = Spectra(name = 'Methane_Line_overview')

#%% 2. add an observer
my_spectrum_overview.observer = Observer(nu_min =  3055.0,    # [1/cm]lower wavenumber
                                         nu_max =  3090.0,    # [1/cm]upper wavenumber
                                )
#%% 3. add gas_cells (may consist of multiple gasses) with gasses
# cell 0
my_spectrum_overview.add_gas_cell()            # number of gasses
my_spectrum_overview.gas_cells[-1].add_gas(**water_cell)
# cell 1
my_spectrum_overview.add_gas_cell()
my_spectrum_overview.gas_cells[-1].add_gas(**methane_cell)

#%% 4. download (running hapi functions)
with HiddenPrints():            # disables hapi stdout
    my_spectrum_overview.download(line_list=True)
```
