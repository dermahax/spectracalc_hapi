```python
%%capture
%load_ext autoreload
%autoreload 2
import os
from pathlib import Path
import matplotlib
import configparser
os.chdir('..')
from classes import Spectra, Observer, HiddenPrints
```

```python
%%capture


VMR_Methan = 500E-6
VMR_Water = 10E-2
### Selektivit채t passiv

#%% 1. create a spectrum
selectivity_passiv = Spectra(name = 'CH4_selectivity_passiv')

#%% 2. add an observer
selectivity_passiv.observer = Observer(unit  = 'lam',      #'wav' : wavenumer [1/cm] | 'lam' : wavelength [nm].
                                lower = 3080.  , #lower =  3260.,    
                                upper = 3530,  #upper =  3260.6,   
                                )
#%% 3. add gas_cells (may consist of multiple gasses) with gasses
# cell 1
selectivity_passiv.add_gas_cell(temperature    = 296,          # K. Default is 296
                         pressure       = 1,            # atm. Default is 1
                         length         = 100,           # cm. Default is 10
                         no_gasses      = 1)            # number of gasses. Default is 1

selectivity_passiv.gas_cells[-1].add_gas(gas_name  = "H2O",    # as in hitran
                                VMR        = VMR_Water)    # E-6 = ppm
# cell 2
selectivity_passiv.add_gas_cell(temperature    = 296,          # K. Default is 296
                         pressure       = 1,            # atm. Default is 1
                         length         = 100,           # cm. Default is 10
                         no_gasses      = 1)            # number of gasses. Default is 1

selectivity_passiv.gas_cells[-1].add_gas(gas_name  = "CH4",    # as in hitran
                                VMR        = VMR_Methan)    # E-6 = ppm 

#%% 4. download (running hapi functions)
selectivity_passiv.download(line_list = False,         # by setting to false, the linelist will nor be downloaded nor plotted.
                     step      = 0.5         # The step size determines the resolution of the graph. It is in the unit of wavenumbers (1/cm).
                    )  


### Selektivit채t aktiv

#%% 1. create a spectrum
selectivity_active = Spectra(name = 'CH4_selectivity_active')

#%% 2. add an observer
selectivity_active.observer = Observer(unit  = 'lam',      #'wav' : wavenumer [1/cm] | 'lam' : wavelength [nm].
                                lower = 3265.  ,    
                                upper = 3275,  
                                )
#%% 3. add gas_cells (may consist of multiple gasses) with gasses

# cell 1
selectivity_active.add_gas_cell(temperature    = 296,          # K. Default is 296
                         pressure       = 1,            # atm. Default is 1
                         length         = 100,           # cm. Default is 10
                         no_gasses      = 1)            # number of gasses. Default is 1

selectivity_active.gas_cells[-1].add_gas(gas_name  = "H2O",    # as in hitran
                                VMR        = VMR_Water)    # E-6 = ppm 

# cell 2
selectivity_active.add_gas_cell(temperature    = 296,          # K. Default is 296
                         pressure       = 1,            # atm. Default is 1
                         length         = 100,           # cm. Default is 10
                         no_gasses      = 1)            # number of gasses. Default is 1

selectivity_active.gas_cells[-1].add_gas(gas_name  = "CH4",    # as in hitran
                                VMR        = VMR_Methan)    # E-6 = ppm 

#%% 4. download (running hapi functions)
selectivity_active.download(line_list = False,         # by setting to false, the linelist will nor be downloaded nor plotted.
                     step      = 0.01         # The step size determines the resolution of the graph. It is in the unit of wavenumbers (1/cm).
                    )  

###############################

### Sensitivit채t passiv

#%% 1. create a spectrum
sensitivity_passiv = Spectra(name = 'CH4_sensitivity_passiv')

#%% 2. add an observer
sensitivity_passiv.observer = Observer(unit  = 'lam',      #'wav' : wavenumer [1/cm] | 'lam' : wavelength [nm].
                                lower = 3080.  , #lower =  3260.,    
                                upper = 3530,  #upper =  3260.6,   
                                )
#%% 3. add gas_cells (may consist of multiple gasses) with gasses

# cell 0
sensitivity_passiv.add_gas_cell(temperature    = 296,          # K. Default is 296
                         pressure       = 1,            # atm. Default is 1
                         length         = 100,           # cm. Default is 10
                         no_gasses      = 1)            # number of gasses. Default is 1

sensitivity_passiv.gas_cells[-1].add_gas(gas_name  = "CH4",    # as in hitran
                                VMR        = VMR_Methan)    # E-6 = ppm 
# cell 0 - 2
sensitivity_passiv.add_gas_cell(temperature    = 296,          # K. Default is 296
                         pressure       = 1,            # atm. Default is 1
                         length         = 100,           # cm. Default is 10
                         no_gasses      = 1)            # number of gasses. Default is 1
sensitivity_passiv.gas_cells[-1].add_gas(gas_name  = "CH4",    # as in hitran
                                VMR        = VMR_Methan)    # E-6 = ppm 

#%% 4. download (running hapi functions)
sensitivity_passiv.download(line_list = False,         # by setting to false, the linelist will nor be downloaded nor plotted.
                     step      = 0.5         # The step size determines the resolution of the graph. It is in the unit of wavenumbers (1/cm).
                    )  



### Sensitivit채t aktiv

#%% 1. create a spectrum
sensitivity_active = Spectra(name = 'CH4_sensitivity_active')

#%% 2. add an observer
sensitivity_active.observer = Observer(unit  = 'lam',      #'wav' : wavenumer [1/cm] | 'lam' : wavelength [nm].
                                lower = 3265.  ,    
                                upper = 3275,  
                                )
#%% 3. add gas_cells (may consist of multiple gasses) with gasses

# cell 0
sensitivity_active.add_gas_cell(temperature    = 296,          # K. Default is 296
                         pressure       = 1,            # atm. Default is 1
                         length         = 100,           # cm. Default is 10
                         no_gasses      = 1)            # number of gasses. Default is 1

sensitivity_active.gas_cells[-1].add_gas(gas_name  = "CH4",    # as in hitran
                                VMR        = VMR_Methan)    # E-6 = ppm 

sensitivity_active.add_gas_cell(temperature    = 296,          # K. Default is 296
                         pressure       = 1,            # atm. Default is 1
                         length         = 100,           # cm. Default is 10
                         no_gasses      = 1)            # number of gasses. Default is 1
sensitivity_active.gas_cells[-1].add_gas(gas_name  = "CH4",    # as in hitran
                                VMR        = VMR_Methan)    # E-6 = ppm 

#%% 4. download (running hapi functions)
sensitivity_active.download(line_list = False,         # by setting to false, the linelist will nor be downloaded nor plotted.
                     step      = 0.01         # The step size determines the resolution of the graph. It is in the unit of wavenumbers (1/cm).
                    )  



```
